Question 5

```
#Histogram of attribute Rain for the ten stations
T= df.groupby('stationid')
station = df.groupby('stationid').groups
for i in station:
  group = T.get_group(i)
  plt.hist(group['rain'], bins = 10)
  plt.title(f'Station {i}') 
  plt.show()

```
